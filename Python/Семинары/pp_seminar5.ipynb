#### Задача 1

Напишите функцию, которая будет считать скорость свободно падающего тела (в момент столкновения с поверхностью) с учетом высоты падения `h` (в метрах). Формула известна: $\sqrt{2gh}$, где `g` - ускорение свободного падения, константное для каждой планеты, а `h` - высота. Пусть `g` по умолчанию принимает значение `9.8`.

Также функция должна иметь возможность возвращать ответ как в `'м/с'`, так и в `'км/ч'` (по запросу пользователя). В случае, если пользователь вводит какие-то другие единицы измерения, должно быть вызвано исключение `ValueError`.

````python
def velocity(h: float, g=9.8, measure='м/с') -> float: # float - просто показываем для стороннего человека
                                    # какой имменно тип мы хотим иметь для h
    # проверим агкумент measure, перед тем как...
    measure_permitted = ['м/с', 'км/ч']
    if measure not in measure_permitted:
        raise ValueError(f'measure argumeте must be one of {measure_permitted}') # вывод сообщения об ошибке
    
    # ... призводить основные вычисления
    result = (2 * g * h) ** (1/2)
    
    if measure != 'м/с':
        result *= 3.6
    
    return(result) # заадем вывод, результат в м/с
````

````python
velocity(1.96, measure='км/ч') # задаем h
````

#### Задача 2
Напишите следующую функцию:

- она должна принимать на вход название текстового файла (путь к нему);
- а также разделитель в этом файле;
- функция должна сохранять эту же таблицу в формате HTML. Должны быть использованы теги \<table>, \<tr>, \<th> и \<td>;
- функция должна возвращать путь к новому файлу.

<table>
    <tr>
        <th>Заголоовок 1 </th>
        <th>Заголоовок 2 </th> 
    </tr>
    <tr>
        <th>ячейка 1 </th>
        <th>ячейка 2 </th> 
    </tr>
    <tr>
        <th>ячейка 3 </th>
        <th>ячейка 4 </th> 
    </tr>
</table>

````python
def from_text_to_html (path: str, sep: str) -> str:
    # procecc file
    file = open(path, 'r') # читаем файл
    text = file.read() # достаем файл
    file.close() # закрывем файл

    lines = text.strip().split('\n') # strip убирает пустые сроки вконце, 
                                        # делим текст на строки

    headers = lines[0].replace('п»ї','').split(sep) # удаляет неныжные объекты в начале, 
                                                        # извлекаем зоголовки

    lists = []

    for line in lines[1:]:
        lists.append(line.split(sep))
        
    # procecc headers
    headers_html = []
    for header in headers: # для всех зоголовков в зоголовках
        header_formatted = f'<th style="border: 1px solid black">{header}</th>' # добавляем через f строку 
        headers_html.append(header_formatted) # добавляем в список
    headers_html = '<tr>' + ''.join(headers_html) + '</tr>'

    # procecc rows
    rows_html = []

    for row in lists:  
        row_html = []
        
        for value in row: # для всех зоголовков в зоголовках
            value_formatted = f'<td style="border: 1px solid black">{value}</td>' # добавляем через f строку 
            row_html.append(value_formatted)
            # добавляем в список
        row_html = '<tr>' + ''.join(row_html) + '</tr>'
        rows_html.append(row_html)
        
    rows_html = ''.join(rows_html)
    table_html = '<table style="border: 1px solid black">' + headers_html + rows_html + '</table>'

    # save final table
    new_path = path.replace('csv', 'html')
    file = open(new_path, 'w') # открываем файл на запись
    file.write(table_html) # записываем содержимое в файл
    file.close()
    
    return new_path
````

#### Задача 3
В Excel существует функция [СЦЕПИТЬ](https://support.microsoft.com/ru-ru/office/%D1%84%D1%83%D0%BD%D0%BA%D1%86%D0%B8%D1%8F-%D1%81%D1%86%D0%B5%D0%BF%D0%B8%D1%82%D1%8C-8f8ae884-2ca8-4f7a-b093-75d702bea31d). Она позволяет объединять две или более текстовые строки в одну.

Напишите функцию, которая будет принимать на вход текстовые колонки (разделенные символом `'\n'`) до тех пор, пока не будет введена пустая строка. Затем эти колонки должны быть "склеены" в одну через пробел и выведены для пользователя.

```
input:
строка1\nстрока2\nстрока3
row1\nrow2\nrow3


output:
строка1 row1
строка2 row3
строка3 row3
```

````python
def excel_join () -> None:
    columns_split = []

    column = input()

    while column != '':
        columns_split.append(column.split('\\n'))
        column = input()

    new_column = []

    for row in zip(*columns_split):
        new_column.append(' '.join(row))
    print(*new_column, sep='\n')
````

````python
excel_join()
````
