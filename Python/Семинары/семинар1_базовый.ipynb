Напишите программу, которая будет принимать на вход любое действительное число, приобразовывать его в соотвествующий тим и печатать его для пользователя (либо 'int', либо 'float') в приобразованном виде
**примечание: пользователь может ввести только целое или действительное число**
1. нужно попросить ввести число("input()"); (1,2 - целое, 5.5, 6.3 - действительное);
2. понять входит ли "." в строку; Можно использовать метод '.find()'
3. преобразовать строку либо в int либо в float использую int() или float()
4. напечатать для пользователя.




```python
number = input()
if number.find('.') != -1: # возвращает индекс или -1,
                            # если символа нет
    result = float(number)
else:
    result = int(number)
print(result)
```

```python
# способ 2
number = input()
if '.' in number: # True/False
    result = float(number)
else:
    result = int(number)
print(result)
```

и печатать его для пользователя, программа должна обращаться к пользователю "введите слова через запятую". 
1. надо попросить пользователя ввести слова через запятую (input)
