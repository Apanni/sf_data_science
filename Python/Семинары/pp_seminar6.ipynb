#### Задача 1
Напишите функцию, которая будет переводить `int` в `str` без использования функции `str()`.

````python
def int1_to_str(number: int) -> str:
    return chr(48 + number)

def int_to_str(number: int) -> str:
    a = abs(number) # по модулю
    
    number_str = ''
    
    while a // 10 > 0:
        number_str += int1_to_str(a % 10)
        a //=10
    else:  # else дает возможность что-то еще сделать пока не вышли из цикла while
        number_str += int1_to_str(a)
    
    number_str = number_str[::-1]

    if number < 0:
        number_str = '-' + number_str
    return number_str
````

Мы уже знаем, как прочитать таблицу в текстовом формате и создать на ее основе красивую таблицу в HTML. Теперь давайте попробуем решить обратную задачу: спарсим [эту](https://meta.wikimedia.org/wiki/List_of_countries_by_regional_classification) таблицу используя `BeautifulSoup` из `bs4` и сохраним в формате tsv (как csv, только через `'\t'`).

````python
# !pip install requests
# !pip install lxml
# !pip install bs4

# table: 69-1812
# <table><thead>headers</thead><tbody>остальное</tbody></table>
````

````python
def from_html_to_text(soup: BeautifulSoup, name: str, sep='\t') -> None:
    table = soup.find('tbody') # ищим тэги, ищет первое вхождение
    all_trs = table.find_all('tr') # можно сделать также table.tr

    # headers
    first_tr = all_trs[0] 
    headers = [th.text.strip() for th in first_tr.find_all('th')]

    # data
    data = [[td.text.strip() for td in tr.find_all('td')] for tr in all_trs[1:]]

    # join all
    lines = [headers] + data
    file_text = '\n'.join([sep.join(line) for line in lines])
    
    file = open(f'{name}.txt', 'w', encoding='utf-8') # если ошибка, нужно задавать кодировку
    file.write(file_text)
    file.close
    
    return f'{name}.txt'
    
````

Предположим, что мы не только хотим прочитать таблицу из файла (или с HTML-страницы), но еще и записать ее в нашу базу данных. Соответственно, эту таблицу необходимо создать. Для этого нужен примерно такой код:

```
CREATE TABLE distributors (
    did integer,
    name varchar(40)
);
```

PostgreSQL [располагает](https://www.postgresql.org/docs/current/datatype.html) следующими типами данных (это не все, что есть, но все, что нам сейчас пригодится):

<table>
    <tr>
        <th>Python</th>
        <th>PostgreSQL</th>
    </tr>
    <tr>
        <td>int</td>
        <td>smallint, integer, bigint</td>
    </tr>
    <tr>
        <td>float</td>
        <td>real</td>
    </tr>
    <tr>
        <td>bool</td>
        <td>boolean</td>
    </tr>
    <tr>
        <td>str</td>
        <td>varchar, text</td>
    </tr>
</table>

Реализуйте соответствующую функцию. Она должна иметь такие аргументы:

- `filename` (строка);
- `sep` (строка, по дефолту для CSV);
- `tablename` (строка, по дефолту соответствует названию файла без расширения);
- `varnames` (либо должны быть предоставлены в виде списка, либо добываются из данных - аргумент `use_first`);
- `use_first` (использовать ли первую строку файла в качестве названий колонок. Логический, по дефолту да);
- `dtypes` (либо должны быть предоставлены в виде списка, либо добываются из данных - аргумент `infer_dtypes`);
- `infer_dtypes` (попытаться ли угадать тип данных из файла. Логический, по дефолту нет).

````python
def infer_sql_from_file(
    filename,
    sep=',',
    tablename=None,
    varnames=None,
    dtypes=None
    ):
    
    # из семинара 5
    # прочитаем файл
    file = open(filename, 'r') # читаем файл
    text = file.read() # достаем файл
    file.close() # закрывем файл

    lines = text.strip().split('\n') # strip убирает пустые сроки вконце, 
                                        # делим текст на строки
    # создадим заголовки в зависимости от переменной
    if varnames is None:
        headers = lines[0] # удаляет неныжные объекты в начале, 
        headers = headers.split(sep) # извлекаем зоголовки
    else:
        headers = varnames.copy()

    lists = []

    for line in lines[1:]:
        lists.append(line.split(sep))
    
    # из семинара 4    
    columns = list(zip(*lists)) # позволяет рассматривать все элементы большого 
            # списка как отдельные значения
    headers_columns = zip(headers, columns)
    super_dict = dict(headers_columns)

# создадим название таблицы
    if tablename is None:
        dot_index = filename.rfind('.')
        tablename = filename[:dot_index]
        
    return tablename, headers

  
````
