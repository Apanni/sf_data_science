#### Задача 1
Напишите код, который бы выполнял операцию транспонирования матрицы. Решите задачу двумя способами.

````python
matrix = [
    [60, 46, 81, 25, 86],  # строка
    [60, 12, 80, 60, 98],
    [31, 90, 35, 87, 60],
    [86, 19, 88, 85, 8]
]

# наш код здесь
````

**способ 1**

````python
matrix_T = [] # создаем пустую матрицу, вкотрую вложим перезаписанну
for index in range(len(matrix[0])): # идем по всем индексам строки 
                                    # (они одинаковой длины, т.к. это матрица), 
                                    # вместо 0 можно ставить и другой идекс
    new_row = [] # создаем новый пустой список
    for row in matrix: # из каждой строки берем J-е значение
        new_row.append(row[index]) # добавляем к списку значения 
    
    matrix_T.append(new_row)
````

**вариант 2**

````python
matrix_T = list(zip(*matrix)) # распоковали список звездочкой (*), \
                            # соединили все списки между собой 
                            # (1й к первому, 2й ко второму и т.д.)
print(matrix_T)
````

````python
# полное решение
file = open(full_path, 'r') # читаем файл
text = file.read() # достаем файл
file.close() # закрывем файл

lines = text.strip().split('\n') # strip убирает пустые сроки вконце, 
                                    # делим текст на строки

headers = lines[0].replace('п»ї','').split(';') # удаляет неныжные объекты в начале, 
                                                    # извлекаем зоголовки

lists = []

for line in lines[1:]:
    lists.append(line.split(';'))

columns = list(zip(*lists)) # позволяет рассматривать все элементы большого 
            # списка как отдельные значения
headers_columns = zip(headers, columns)
super_dict = dict(headers_columns)

````

````python
aligments_unique = set(super_dict['Alignment']) # множество set 
````

````python
for value in aligments_unique:
    value_count = super_dict['Alignment'].count(value)
    
    print(f'For value {value} count is {value_count}')
````
